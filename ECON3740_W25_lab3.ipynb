## Group by

Summary statistics by group

Collapsing the data by group is the easiest way of getting stats for continuous vars


```python
# Compute mean of 'AGE_6' grouped by 'SEX'

data.groupby('female').agg(MEAN=('wage', 'mean'))
```

# Regression

## Simple regression

```python
import statsmodels.formula.api as smf

# Create the model
model = smf.ols('wage ~ exper', data)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
```

## Multiple Linear

```python
# Create the model
model1 = smf.ols('wage ~ exper + tenure', data)

# Fit the model
results1 = model1.fit()

# Display regression results
print(results1.summary())
```

## Log

```python
# Create the model
model2 = smf.ols('wage ~ exper', data)

# Fit the model
results2 = model2.fit()

# Display regression results
print(results2.summary())
```

```python
import numpy as np

# Create the model
model2 = smf.ols('wage ~ np.log(exper)', data)

# Fit the model
results2 = model2.fit()

# Display regression results
print(results2.summary())
```

## Categorical variables

```python
# Create the model
model3 = smf.ols('wage ~ C(female)', data)

# Fit the model
results3 = model3.fit()

# Display regression results
print(results3.summary())
```

## Plot regression line

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x='wage',
            y='exper',
            data=data,
            ci=None,
            scatter_kws={'facecolors':'white',
                         'edgecolors':'black'})

plt.show()
```

```python
data['log_wage'] = np.log(data['wage'])

sns.regplot(y='log_wage',
            x='exper',
            data=data,
            ci=None,
            scatter_kws={'facecolors':'white',
                         'edgecolors':'black'})

plt.show()
```
